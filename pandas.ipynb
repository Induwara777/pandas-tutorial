If we need some column, There are two method to do.

*  1.  without [ ] == use `loc`
*  2.  with [ ] == `don't use loc`

* Without [  ]

```python
olim.loc[olim["height_cm"] > 215 , ["height_cm","name","born_country"]]
```

* with [  ]

No need use `loc function`

If there is `one column` = It include in ` " " `

If there are `more than one column` = all column names are `entered in [ ].`

This [ ] method named as `binary syntax`

```python
olim[olim["height_cm"] > 215][["name","height_cm","born_country"]]
```

* When there are `more than one condition` to filter, we can use `&` / `|` symbole and `each condition are in () `.Also we can use this symbole in string..Ex `Keith|Patrick`

```python
# use `&`
olim[ (olim["height_cm"] > 215) & (olim["born_country"] =="USA" )]
```

```python
# use `|`
olim[ (olim["height_cm"] > 215) | (olim["born_country"] =="USA" )]
```

3) Getting `interploted values` == `Best one` for this

In pandas, the interpolate() function is used to fill NaN (Not a Number) values in a DataFrame or Series. It provides several methods to interpolate missing values based on different strategies:

1) `Linear Interpolation` : This method fills NaN values using a linear interpolation between the non-missing values that precede and follow each NaN.Default version

2) `Time-based Interpolation` : If your index is a datetime index, you can use methods like time, index, or values to interpolate based on time.

3) `Polynomial Interpolation` : This method fits a polynomial function to the non-missing values and uses it to estimate the NaN values.

4) `Nearest Values` : This method fills NaN values with the nearest non-missing value.

5) `Spline Interpolation` : This uses cubic spline interpolation to fill NaN values.
